```python
import pandas as pd
```

```python
param_df = pd.read_csv('result_of_finding_best_params_for_sgd.csv')
mean_df = param_df.groupby(['n_iter', 'n_factors'])[['train_rmse', 'test_rmse']].mean().reset_index()
```

```python
mean_df
```

```python
%matplotlib inline
import matplotlib.pyplot as plt
import seaborn as sns
sns.set()

factors = mean_df['n_factors'].unique()
for factor in factors:
    df = mean_df[mean_df['n_factors'] == factor]
    plt.plot(df['n_iter'], df['test_rmse'], label=f'n_factors={factor}')
plt.legend()
plt.xlabel('n_iter')
plt.ylabel('mean of 30 test_rmse')
plt.title('mean of 30 test_rmse for different n_iter and n_factors')
plt.savefig('mean_of_30_test_rmse_for_different_n_iter_and_n_factors.png')
plt.show()

```
